```python
import pandas as pd
import scipy.stats as stats
from IPython.display import display

# Assuming df_cat is your DataFrame and 'UC' is the dependent variable

# List of independent variables (all columns except 'UC')
independent_columns = [col for col in df_cat.columns if col != 'UC']

# Create an empty list to store the results
results = []

# Perform Chi-Square test for each independent variable with 'UC' as the dependent variable
for col in independent_columns:
    # Create a contingency table between 'UC' and the independent variable
    contingency_table = pd.crosstab(df_cat['UC'], df_cat[col])
    
    # Perform Chi-Square test
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    
    # Check if the p-value is below the significance level (e.g., 0.05)
    significance = 'Significant' if p < 0.05 else 'Not Significant'
    
    # Append results to the list
    results.append([col, 'UC', p, significance])



# Convert the results into a DataFrame
results_df = pd.DataFrame(results, columns=['Independent Variable', 'Target Variable', 'P-Value', 'Significance'])

# Display the results table
# print(results_df)

# Make the table more publication-ready with styling
styled_results_df = results_df.style.set_table_styles(
    [
        {'selector': 'th', 'props': [('background-color', '#f2f2f2'), ('font-weight', 'bold'), ('color', 'black')]},
        {'selector': 'td', 'props': [('padding', '8px'), ('text-align', 'right'), ('color', 'black')]},
        {'selector': 'table', 'props': [('border-collapse', 'collapse'), ('width', '100%')]},
        {'selector': 'tr:nth-child(even)', 'props': [('background-color', '#f9f9f9')]},
        {'selector': 'tr:nth-child(odd)', 'props': [('background-color', 'white')]},
    ]
).format({'p-value': '{:.2e}'})  # Format p-values in scientific notation

# Display the styled DataFrame
styled_results_df

```

```python
import pandas as pd
import statsmodels.api as sm

X = df[['Age ', 'Sex', 'Current Smoker', 'IBD-U', '5-ASA', 'Azathioprine or 6-MP', 
        'Adalimumab', 'vedolizumab', 'current steroid use', 'Tofacitinib', 'Ustekinumab', 
       'New Previous surgery', 'New previous biologic used']]  # Select your predictor columns
y = df['UC']  # Target variable

# Add constant term to the feature matrix (intercept term)
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.Logit(y, X)
result = model.fit()

# Display the summary of the regression
print(result.summary())
```

```python
# Make predictions on the test set
y_pred_prob = result.predict(X)  # Predicting probabilities

# Convert probabilities to binary outcomes based on a threshold of 0.5
y_pred = (y_pred_prob > 0.5).astype(int)
```

### Accuracy - Confusion Matrix

```python

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y, y_pred))
print()
print(classification_report(y, y_pred))
print()
print(f"Accuracy: {accuracy_score(y, y_pred):.4f}")
```

```python
# Create necessary components
coefficients = result.params
odds_ratios = np.exp(result.params)
conf = result.conf_int(alpha=0.05)
p_values = result.pvalues

# Combine the confidence intervals into a single string (lower, upper bounds)
conf_intervals = conf.apply(lambda row: f"({row[0]:.3f}, {row[1]:.3f})", axis=1)

# Create the combined table
summary_table = pd.DataFrame({
    'Coefficient (log-odds)': coefficients,
    'Odds Ratio': odds_ratios,
    '95% Confidence Interval': conf_intervals,
    'P-value': p_values
})

# Round the values for better readability
summary_table = summary_table.round({'Coefficient (log-odds)': 3, 'Odds Ratio': 3, 'P-value': 3})
print(summary_table)
```
